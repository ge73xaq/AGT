```python
%load_ext autoreload
%autoreload 2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
# import uniform distribution
from scipy.stats import uniform
from scipy.stats import binom
from scipy.stats import norm
import scipy.special as sc
import scipy.optimize
from scipy.optimize import fsolve
import time
from scipy.stats import truncnorm
np.random.seed(5)
```

```python
def auction_exp(data_auction, p):
    """Train a normalizing flow model with maximum likelihood.

    Args:
        model: normalizing flow model. Flow or StackedFlows
        dataset: dataset containing data to fit. Dataset
        batch_size: number of samples per batch. int
        max_epochs: number of training epochs. int
        frequency: frequency for plotting density visualization. int
        
    Return:
        model: trained model. Flow or StackedFlows
        losses: loss evolution during training. list of floats
    """
    revenue = 0
    for bid in data_auction:
        if bid >= p:
          #change the bidder with price p
          revenue += p
    
    revenue = revenue/len(data_auction)
    return revenue
```

```python
def rejection_sampling(pi1, pi2):

    # def f(x):
    #   return [x[0] * (1 + np.log(x[1]/x[0])) - u,
    #           x[0] *(2*x[1]-x[0]) - u**2 - sigma**2]


    # root = fsolve(f, [ini1,ini2])

    # pi1 = min(root)
    # pi2 = max(root)
    #using rejection sampling, get N random price samples
    N = 10000
    n = 0
    #A =(pi2/pi1-1)*(pi2-pi1)/don
    don = pi2*np.log(pi2/pi1) - (pi2-pi1)
    env =(pi2/pi1-1)/don
    rej_sample =[]
    while(n <= N):
        x = uniform.rvs(loc=pi1, scale=pi2-pi1)
        a = (pi2/x-1)/don/env
        u1 = uniform.rvs()
        if (u1 <= a):
            rej_sample.append(x)
            n += 1
    return rej_sample
```

```python
# Deterministic multiple items experiment on truncated normal distribution:
sigma_list = []

r_list =[]
upperbound =[]
p_d_list = []
rev_tnorm_sep=[]
rev_tnorm_bund=[]

p_opt_list =[]

def Opt_p(x):
    return x-((1-k*(norm.cdf(x, loc = u,scale = sigma)-norm.cdf(0, loc = u,scale = sigma)))/(k*norm.pdf(x, loc = u,scale = sigma)))

def f_int(x):
    return (x * (1 + np.log(k2/(x**2) + 1) - np.log(2)) -mean )


def F(x):
    return ((x-1)**3)/((2*x-1)**2) - r**2   


def tnorm_parmratio(x):
    return ((x+(norm.pdf(-x)/(1-norm.cdf(-x))))**2/(1-x*(norm.pdf(-x)/(1-norm.cdf(-x)))-
                                                    (norm.pdf(-x)/(1-norm.cdf(-x)))**2)-mean**2/var) 

APX_tnorm_list = []

step =np.linspace(0, 8,50)
m = 5
n = 100000
myclip_a = 0
myclip_b = np.inf
u = 1
tic = time.clock()

for i in step:
    u = 1
    sigma = 1 + i
    #computer alpha beta for truncated normal distribution
    a, b = (myclip_a - u) / sigma, myclip_b

    #constant of truncated normal distribution
    k = 1/(1-norm.cdf(0, loc = u,scale = sigma))
    #compute the truncated normal mean and variance
    mean, var = truncnorm.stats(a , b ,loc = u,scale = sigma, moments='mv')

    r = np.sqrt(var)/mean
    r_list.append(r)
    sigma_list.append(np.sqrt(var))
    p_opt = fsolve(Opt_p, mean)
    print(Opt_p(p_opt))

    #now we simulate selling separately mechanism 
    bid_tnorm = truncnorm.rvs(a , b , loc = u,scale = sigma,size=n)
    opt = auction_exp(bid_tnorm,p_opt)
    rev_tnorm_sep.append(opt[0]*m)
    
    #now we simulate bundle selling mechanism 
    mean = m*mean
    upperbound.append(mean)
    var = m*var
    r = np.sqrt(var)/mean
    p_d = fsolve(F, [mean])
    
    #reserve price proposed by the Robust paper
    p_reserve = p_d*mean/(2*p_d-1)
    print(F(p_d))
    bid_bundle=[]
    j=0
    while j < n:
        #we generate m random variables from uniform distribution [a,b]
        #the sum of these variables denote one auction bid, we will simulate n auctions,
        #thus require n convoluted random bids
        bid_bundle.append(sum(truncnorm.rvs(a , b , loc = u,scale = sigma,size=m))) 
        j += 1
    
    

    rev_tnorm_bund.append(auction_exp(bid_bundle,p_reserve)[0])
    


toc = time.clock()
print('time: ' + str(toc-tic))
print(rev_tnorm_sep)
print(rev_tnorm_bund)
```

```python
# Randomisation multiple item experiment compute the log-lottery with truncated normal distribution:
sigma_list = []
DAPX_tnorm = []
r_list =[]
p_opt1_list = []
p_d_list = []
APX_tnorm_bulk = []
p_opt_list =[]

def Opt_p(x):
    return x-((1-k*(norm.cdf(x, loc = u,scale = sigma)-norm.cdf(0, loc = u,scale = sigma)))/(k*norm.pdf(x, loc = u,scale = sigma)))

def f_int(x):
    return (x * (1 + np.log(k2/(x**2) + 1) - np.log(2)) -mean )
 
def tnorm_parmratio(x):
    return ((x+(norm.pdf(-x)/(1-norm.cdf(-x))))**2/(1-x*(norm.pdf(-x)/(1-norm.cdf(-x)))-
                                                    (norm.pdf(-x)/(1-norm.cdf(-x)))**2)-mean**2/var) 

APX_tnorm_list = []

step =np.linspace(0, 6,20)
m = 10
n = 10000
myclip_a = 0
myclip_b = np.inf
u = 1
tic = time.clock()

for i in step:
    u = 1
    sigma = 1 + i
    #computer alpha beta for truncated normal distribution
    a, b = (myclip_a - u) / sigma, myclip_b

    #constant of truncated normal distribution
    k = 1/(1-norm.cdf(0, loc = u,scale = sigma))
    mean, var = truncnorm.stats(a , b ,loc = u,scale = sigma, moments='mv')

    #use truncated normal mean and variance to compute second moment
    k2 = mean**2 + var

    r = np.sqrt(var)/mean
    r_list.append(r)
    sigma_list.append(np.sqrt(var))
    p_opt = fsolve(Opt_p, mean)
    p_opt_list.append(p_opt)
    print(Opt_p(p_opt))

    pi1 = fsolve(f_int, [mean])
    pi2 = 0.5*(k2/pi1 +pi1)  
    #print(mean, var)
    #print(pi1,pi2)
    opt_list=[]
    revenue_list=[]
    for j in np.arange(0, m): #for each item, run a auction simulation n times
        bid_tnorm = truncnorm.rvs(a , b , loc = u,scale = sigma,size=n)
        opt = auction_exp(bid_tnorm,p_opt)
        opt_list.append(opt)
        rexp_revenue_list=[]
        rej_sample=[]
        rej_sample = rejection_sampling(pi1,pi2)
        for rp_reserve in rej_sample:
            rexp_revenue = auction_exp(bid_tnorm,rp_reserve)
            rexp_revenue_list.append(rexp_revenue)
        revenue_list.append(np.average(rexp_revenue_list))
        
    APX_tnorm_list.append((sum(opt_list)/sum(revenue_list))) 
    
    
    mean = m*mean
    var = m*var
    #print(mean, var)
    r = np.sqrt(var)/mean
    ratio = fsolve(tnorm_parmratio,1/r)
    u = mean/(1+ (1/ratio)*norm.pdf(-ratio)/(1-norm.cdf(-ratio)))
    sigma = u/ratio
    a, b = (myclip_a - u) / sigma, myclip_b
    print(tnorm_parmratio(ratio))
    print(truncnorm.stats(a , b ,loc = u,scale = sigma, moments='mv'))
    bid_tnorm = truncnorm.rvs(a , b , loc = u,scale = sigma,size=n)
    p_opt1 = fsolve(Opt_p, mean)
    print(Opt_p(p_opt1))
    p_opt1_list.append(p_opt)
    opt = auction_exp(bid_tnorm,p_opt1)
    
    #bulk randomization sell
    k2 = mean**2 + var
    pi1 = fsolve(f_int, [mean])
    pi2 = 0.5*(k2/pi1 +pi1)
    #print(pi1, pi2)
    rej_sample = []
    rexp_revenue_bulk=[]
    rej_sample = rejection_sampling(pi1,pi2)

    for rp_reserve in rej_sample:
        rexp_revenue = auction_exp(bid_tnorm,rp_reserve)
        rexp_revenue_bulk.append(rexp_revenue)

    APX_tnorm_bulk.append((opt/np.average(rexp_revenue_bulk))) 
    


toc = time.clock()
print('time: ' + str(toc-tic))
print(APX_tnorm_list)
print(APX_tnorm_bulk)
```
