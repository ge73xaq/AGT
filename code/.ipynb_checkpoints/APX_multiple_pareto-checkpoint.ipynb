```python
%load_ext autoreload
%autoreload 2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
# import uniform distribution
from scipy.stats import uniform
from scipy.stats import binom
from scipy.stats import norm
import scipy.special as sc
import scipy.optimize
from scipy.optimize import fsolve
from scipy.stats import pareto
import time
np.random.seed(5)
```

```python
def auction_exp(data_auction, p):
    """Train a normalizing flow model with maximum likelihood.

    Args:
        model: normalizing flow model. Flow or StackedFlows
        dataset: dataset containing data to fit. Dataset
        batch_size: number of samples per batch. int
        max_epochs: number of training epochs. int
        frequency: frequency for plotting density visualization. int
        
    Return:
        model: trained model. Flow or StackedFlows
        losses: loss evolution during training. list of floats
    """
    revenue = 0
    for bid in data_auction:
        if bid >= p:
          #change the bidder with price p
          revenue += p
    
    revenue = revenue/len(data_auction)
    return revenue
```

```python
def rejection_sampling(pi1, pi2):

  # def f(x):
  #   return [x[0] * (1 + np.log(x[1]/x[0])) - u,
  #           x[0] *(2*x[1]-x[0]) - u**2 - sigma**2]


  # root = fsolve(f, [ini1,ini2])
    
  # pi1 = min(root)
  # pi2 = max(root)
  #using rejection sampling, get N random price samples
  N = 10000
  n = 0
  #A =(pi2/pi1-1)*(pi2-pi1)/don
  don = pi2*np.log(pi2/pi1) - (pi2-pi1)
  env =(pi2/pi1-1)/don
  rej_sample =[]
  while(n <= N):
    x = uniform.rvs(loc=pi1, scale=pi2-pi1)
    a = (pi2/x-1)/don/env
    u1 = uniform.rvs()
    if (u1 <= a):
      rej_sample.append(x)
      n += 1
  return rej_sample
```

```python
# Experiment Deterministic expected revenue for pareto(1) distribution:
sigma_list = []
r_list =[]
reserve_p_list = []
p_d_list = []

c_list=[]
tic = time.clock()
#m is the number of items
m = 5

rev_pareto_sep = []
rev_pareto_bund = []
step = np.linspace(1,20,50)
n = 100000

def F(x):
    return ((x-1)**3)/((2*x-1)**2) - r**2   

def f(x):
    return (x * (1 + np.log(k2/(x**2) + 1) - np.log(2)) -u )
 

for i in step:
    r = i
    c = 1+np.sqrt(1+1/(r**2))
    c_list.append(c)
    u, var = pareto.stats(c, moments='mv')
    sigma = np.sqrt(var)
    bid_pareto= pareto.rvs(c,size=n)
    #print(u,sigma, r,c)
    r = sigma/u
    r_list.append(r)

    opt = 1
  
    rev_pareto_sep.append(opt*m) 

    u = m*u
    sigma = np.sqrt(m)*sigma
    r = sigma/u
    
    p_d = fsolve(F, [u])
    
    #reserve price proposed by the Robust paper
    p_reserve = p_d*u/(2*p_d-1)
    print(F(p_d))
    bid_bundle=[]
    j=0
    while j < n:
        #we generate m random variables from uniform distribution [a,b]
        #the sum of these variables denote one auction bid, we will simulate n auctions,
        #thus require n convoluted random bids
        bid_bundle.append(sum(pareto.rvs(c ,size=m))) 
        j += 1

    rev_pareto_bund.append(auction_exp(bid_bundle,p_reserve)[0])
    

toc = time.clock()
print('time: ' + str(toc-tic))
print(rev_pareto_sep)
print(rev_pareto_bund)
```

```python
plt.figure(figsize=(15,10))
plt.xlabel("r = \u03C3/u",fontsize=16)
plt.ylabel("REV",fontsize=16)
plt.tick_params(axis='x', labelsize= 14)
plt.tick_params(axis='y', labelsize= 14)
plt.plot(r_list, rev_pareto_bund,marker="o",color="seagreen",label="bundle sell")
plt.plot(r_list, rev_pareto_sep,marker="o",color="#0065bd",label="separate sell")
# lower_bound = 1+np.log(1+np.power(r_list,2))
# #plt.plot(r_list, ratio,color="blue",label="worst case rho_d")
# plt.plot(r_list, lower_bound,color="skyblue",label="lower bound 1+ln(1+r^2) ")
plt.legend(loc="upper left",fontsize = 16)
plt.savefig("optmulPareto.png",bbox_inches='tight')
plt.show()
```

```python
# Experiment Deterministic expected revenue for pareto(0) distribution:

r_list0 =[]
c_list=[]
tic = time.clock()
#m is the number of items
m = 5

rev_pareto_sep0 = []
rev_pareto_bund0 = []
step = np.linspace(1.0001, 10, 50)
n = 100000

def F(x):
    return ((x-1)**3)/((2*x-1)**2) - r**2   

def f(x):
    return (x * (1 + np.log(k2/(x**2) + 1) - np.log(2)) -u )
 

for i in step:
 
    c = 2*i**2/(i**2-1)
    c_list.append(c)
    u, var = pareto.stats(c, moments='mv')
    sigma = np.sqrt(var)
    bid_pareto= pareto.rvs(c,size=n)
    #print(u,sigma, r,c)
    u = u-1
    r = sigma/u
    r_list0.append(r)

    p_opt = 1/(c-1)

    rev_pareto_sep0.append(auction_exp(bid_pareto-1,p_opt)*m) 

    u = m*u
    sigma = np.sqrt(m)*sigma
    r = sigma/u
    
    p_d = fsolve(F, [u+0.5])
    
    #reserve price proposed by the Robust paper
    p_reserve = p_d*u/(2*p_d-1)
    print(F(p_d))
    bid_bundle=[]
    j=0
    while j < n:
        #we generate m random variables from uniform distribution [a,b]
        #the sum of these variables denote one auction bid, we will simulate n auctions,
        #thus require n convoluted random bids
        bid = pareto.rvs(c ,size=m)-1
        bid_bundle.append(sum(bid)) 
        j += 1

    rev_pareto_bund0.append(auction_exp(bid_bundle,p_reserve)[0])
    

toc = time.clock()
print('time: ' + str(toc-tic))
print(rev_pareto_sep0)
print(rev_pareto_bund0)
```
